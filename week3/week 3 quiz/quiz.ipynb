## Question 4

```python
df = pd.read_csv('data_coffeeshop.csv', header="infer")
df.head()
```

```python
# .1   Make a frequency table of 'yearOfStart' and visualize by year.
# ‚óè   Sort by the year.
# ‚óè   Draw a line plot from 1997 to 2014.
years_table = df['Year of Start'].value_counts().sort_index()
plt.xlim(1997,2014)
plt.plot(years_table.index, years_table.values, '--')
plt.show()
```

```python
state_in = df[df['Current State'] == 'In']['Year of Start'].value_counts().sort_index()
state_out = df[df['Current State'] == 'Out']['Year of Start'].value_counts().sort_index()
plt.xlim(1997,2014)
plt.plot(state_in.index, state_in.values, label='In')
plt.plot(state_out.index, state_out.values, '--', label='Out')
plt.legend()
plt.show()
```

```python
df.sort_values(by='Year of Start', inplace=True)
fig = plt.figure()
df_in = df[df['Current State'] == 'In']
df_out = df[df['Current State'] == 'Out']

plt.plot(df_in['Year of Start'], label='In')
plt.plot(df_out['Year of Start'], label='Out')
plt.xlim(1997, 2014)
plt.legend()
plt.show()
```
